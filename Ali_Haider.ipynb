```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("HR_Analytics.csv")
```

```python
dataset.head()
```

```python
dataset.dropna(how='any',inplace = True)
```

```python
dataset.describe().T
```

```python
print(f'Total Rows: {dataset.shape[0]}\n Total Columns: {dataset.shape[1]}')
```

```python
dataset['AgeGroup'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Education', y='MonthlyIncome', data=dataset, palette='Set2')
plt.title('Monthly Income by Education')
plt.xlabel("Education Level")
plt.ylabel("Monthly Income")
plt.show()
```

```python
plt.scatter(dataset['Age'], dataset['MonthlyRate'], c=dataset['JobLevel'].astype('category').cat.codes)
plt.xlabel('Age')
plt.ylabel('Daily Rate')
plt.title('Scatter Plot of Daily Rate vs Age')
plt.colorbar(label='JobLevel')
plt.show()
```

```python
sns.scatterplot(data=dataset, x='Age', y='MonthlyRate', hue='JobLevel')

plt.show()
```

```python
plt.hist(dataset['AgeGroup'], bins = 5, color='skyblue', edgecolor='black')

plt.title('Distribution of Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

```python
sns.countplot(x='AgeGroup', data=dataset, palette='coolwarm')

plt.title('Distribution of Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

```python
plt.plot(dataset['Age'], dataset['DistanceFromHome'], marker='o', linestyle='-', color='b', label='Distance From Home')

# Add title and labels
plt.title('Stock Prices Over Time', fontsize=14)
plt.xlabel('Time (Days)', fontsize=12)
plt.ylabel('Stock Price (USD)', fontsize=12)

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

plt.show()
```
